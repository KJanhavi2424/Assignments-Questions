```python
#Rename
df.rename(columns={'R&D Spend':'rds','Administration':'ad','Marketing Spend':'ms','State':'st','Profit':'prf'},inplace=True)
df
```

```python
#Model Building
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols('rds~ad+ms+st+prf',data=df).fit()
model.pvalues, model.tvalues
```

```python
#r2, r2_adj
model.rsquared, model.rsquared_adj
```

```python
#rds
rds = smf.ols('ad~rds+ms+prf',data=df).fit()
rds.pvalues, rds.tvalues
```

```python
rds.rsquared, rds.rsquared_adj
```

```python
#Subset Selection
rds.aic
```

```python
#ad
ad = smf.ols('rds~ad+ms+prf',data=df).fit()
ad.pvalues, ad.tvalues
```

```python
ad.rsquared, ad.rsquared_adj
```

```python
ad.aic
```

```python
#ms
ms = smf.ols('prf~ms+ad+rds',data=df).fit()
ms.pvalues, ms.tvalues
```

```python
ms.rsquared, ms.rsquared_adj
```

```python
ms.aic
```

```python
#st
st = smf.ols('ms~st+ad+prf',data=df).fit()
st.pvalues, st.tvalues
```

```python
st.rsquared, st.rsquared_adj
```

```python
st.aic
```

```python
#prf
prf = smf.ols('ms~prf+rds+ad',data=df).fit()
prf.pvalues, prf.tvalues
```

```python
prf.rsquared, prf.rsquared_adj
```

```python
prf.aic
```
