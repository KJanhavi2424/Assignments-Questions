# Set3 - Q5. Mozilla Firefox...

```python
#Importing necessary librariries
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
#Apply One-sample One-tail test
```

```python
z_scores=(0.046-0.05)/(np.sqrt((0.05*(1-0.05))/2000))
z_scores
```

```python
#Find probability assuming Ho, so as to compare with Type-1 error=0.05
```

```python
p_value= 1-stats.norm.cdf(abs(z_scores))
p_value
```
