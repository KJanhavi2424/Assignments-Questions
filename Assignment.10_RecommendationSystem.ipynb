```python
#Importing Necessary Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#Read The Dataset
df = pd.read_csv('https://raw.githubusercontent.com/KJanhavi2424/My-Datasets/main/book.csv',encoding='Latin1')
df
```

```python
df2 = df.drop(columns=['Unnamed: 0'])
df2
```

```python
#EDA And Model Building
#Data Description
df2.describe()
```

```python
df2=df.iloc[:,1:]
df2
```

```python
# Sort by User IDs
df2.sort_values(['User.ID'])
```

```python
#Checking unique UserId and Book.Rating,Book.Title
df2['User.ID'].nunique()
```

```python
df2['Book.Rating'].nunique()
```

```python
df2['Book.Title'].nunique()
```

```python
df2.duplicated().sum()
```

```python
# converting long data into wide data using pivot table
df3=df2.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
df3
```

```python
# Replacing the index values by unique user Ids
df3.index=df2['User.ID'].unique()
df3
```

```python
# Impute those NaNs with 0 values
df3.fillna(0,inplace=True)
df3
```

```python
#Calculate the distance and similarity
from sklearn.metrics import pairwise_distances
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(df3.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
# Set the index and column names to user ids
user_sim2.index=df2['User.ID'].unique()
user_sim2.columns=df2['User.ID'].unique()
user_sim2
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
# Most Similar Users
user_sim2.idxmax(axis=1)
```
