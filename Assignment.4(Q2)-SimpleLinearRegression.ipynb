```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols('YearsExperience~Salary',data=df).fit()
```

```python
#ModelParameters
model.params
```

```python
df.head()
```

```python
#Print values for tvalue and pvalue
model.pvalues , model.tvalues
```

```python
model.tvalues
```

```python
#r2 values
model.rsquared
```

```python
model.rsquared_adj
```
