```python
#A and B manager wants to determine where there is any significant difference in the diameter of the Cutlet between two units.
```

```python
#Importing Necessary Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 300
sns.set_theme(style='darkgrid',palette='rainbow')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#Load The dataset
df = pd.read_csv('https://raw.githubusercontent.com/KJanhavi2424/My-Datasets/main/Cutlets.csv')
df
```

```python
df.head()
```

```python
df.describe()
```

```python
UnitB=pd.Series(df.iloc[:,1])
UnitB
```

```python
UnitA=pd.Series(df.iloc[:,0])
UnitA
```

```python
#2Sample 2tailTest : stats.ttest_ind(array1,array2)
```

```python
#ind = Independent Samples
```

```python
p_value=stats.ttest_ind(UnitA,UnitB)
p_value
```

```python
p_value[1]
```

```python
#2Tail Probability
```

```python
#Compairing p_value with alpha = 0.05(At 5% significance level)
```

```python
p_value=1-stats.t.cdf(p_value[1],df=72)
p_value
```

```python
alpha = 0.05
if p_value <alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```
