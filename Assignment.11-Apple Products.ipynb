```python
#Word Cloud
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
```

```python
df
```

```python
# Define a function to plot word cloud
def plot_cloud(WordCloud):
    plt.figure(figsize=(40,30))
    plt.imshow(wordcloud)
    plt.axis('off')

# Generate Word Cloud

STOPWORDS.add('pron')
STOPWORDS.add('rt')
STOPWORDS.add('yeah')
wordcloud=WordCloud(width=3000,height=2000,background_color='black',max_words=50,
                   colormap='Set1',stopwords=STOPWORDS).generate('Amazing product..it has quality written all over it..great battery, easy to adapt if you are moving from windows..the trackpad, gestures, everything is so damm goodHad read many reviews about the display being bad...it is..if you are trying to look at atomic level details..its a fantastic screen with great audio output..Everything g screams quality and premium about the macbook..love it.')
plot_cloud(wordcloud)
```

```python
# Emotion Mining
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk import word_tokenize
text_tokens=word_tokenize('Amazing product..it has quality written all over it..great battery, easy to adapt if you are moving from windows..the trackpad, gestures, everything is so damm goodHad read many reviews about the display being bad...it is..if you are trying to look at atomic level details..its a fantastic screen with great audio output..Everything g screams quality and premium about the macbook..love it.')
print(text_tokens[0:50])
```

```python
len(text_tokens)
```

```python
# Remove stopwords
from nltk.corpus import stopwords
my_stop_words=stopwords.words('english')

sw_list=['I','The','It','A']
my_stop_words.extend(sw_list)

no_stop_tokens=[word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens)
```

```python
lemmas=[token.lemma_ for token in df]
print(lemmas)
```

```python
#Feature Extraction using Count Vectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
dfcv=cv.fit_transform(lemmas)
```

```python
print(cv.vocabulary_)
```

```python
print(dfcv.toarray()[150:300])
```

```python
print(dfcv.toarray().shape)
```

```python
# CountVectorizer with N-grams (Bigrams & Trigrams)
cv_ngram_range=CountVectorizer(analyzer='word',ngram_range=(1,3),max_features=100)
bow_matrix_ngram=cv_ngram_range.fit_transform(lemmas)
```

```python
print(cv_ngram_range.get_feature_names_out())
print(bow_matrix_ngram.toarray())
```

```python
# TF-IDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfv_ngram_max_features=TfidfVectorizer(norm='l2',analyzer='word',ngram_range=(1,3),max_features=500)
tfidf_matrix_ngram=tfidfv_ngram_max_features.fit_transform(lemmas)
```

```python
print(tfidfv_ngram_max_features.get_feature_names_out())
print(tfidf_matrix_ngram.toarray())
```
