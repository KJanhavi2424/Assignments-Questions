# Set 3

Question.5-In January 2005, a company that monitors Internet Traffic

```python
#Importing Necessary libraries
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
#Apply One-sample one-tail t-test
z_scores=(0.046-0.05)/(np.sqrt((0.05*(1-0.05))/2000))
z_scores
```

```python
#Find probability assuming null hypothesis,so as to compare with Type 1 error= 0.05
p_value= 1-stats.norm.cdf(abs(z_scores))
p_value
```
